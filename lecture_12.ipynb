```python
f = open('.\data\weather.csv')

data = csv.reader(f)
header = next(data)
max_wind = 0
for row in data:
    if(row[3] == ''):
        row[3] = 0
    elif(float(row[3]) > max_wind):
         max_wind = float(row[3])
print('Maximum wind speed is {}.'.format(max_wind))

f.close()
```

```python
# 데이터 합치기
import pandas as pd

df1 = pd.DataFrame({'A':['a10', 'a11','a12'],
'B': ['b10','b11','b12'],
'C': ['c10', 'c11','c12']})

df2 = pd.DataFrame({'B':['b10', 'b11','b12'],
'C': ['c10','c11','c12'],
'D': ['d10', 'd11','d12']})

print(df1)
print(df2) # 이것들을 머지(합치기) 하자.
```

```python
df3 = pd.concat([df1, df2], axis =0, join = 'outer') #컨켓으로 합치기
#(합칠 데이타프레임, axis = 0 이면 아래(행)로 붙이고 1이면 옆(열)으로 붙임, join = 'outer' 은 합집합, 'inner' 이면 교집합으로 생성)
print(df3)
```

```python
df4 = pd.concat([df1, df2], axis =1, join = 'inner')
df4
```

```python
import pandas as pd

df1 = pd.DataFrame({'A':['a10', 'a11','a12'],
'B': ['b10','b11','b12'],
'C': ['c10', 'c11','c12']})

df2 = pd.DataFrame({'B':['b20', 'b21','b22'],
'C': ['c20','c21','c22'],
'D': ['d20', 'd21','d22']})

print(df1)
print(df2) # 이것들을 머지(합치기) 하자.
```

```python
df3 = df1.merge(df2, how = 'outer', on = 'B')
df3
```

```python
# 인덱스를 키로 활용하여 merge()

df1 = pd.DataFrame({'A':['a10', 'a11','a12'],
'B': ['b10','b11','b12'],
'C': ['c10', 'c11','c12']}, index=[0,1,2])

df2 = pd.DataFrame({'B':['b20', 'b21','b22'],
'C': ['c20','c21','c22'],
'D': ['d20', 'd21','d22']}, index=[2,3,4])

df3 = df1.merge(df2, how = 'outer', left_index=True, right_index= True)
df3
```
