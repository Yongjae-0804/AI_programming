```python
import cv2

img1 = cv2.imread('./data/img1.png', 1) #cv2.IMREAD_COLOR
mask = cv2.imread('./data/mask_circle.png', 1) #mask image
img1 = cv2.resize(img1, (300, 300))
mask = cv2.resize(mask, (300, 300))

mask_and = cv2.bitwise_and(img1, mask)
mask_or = cv2.bitwise_or(img1, mask)
mask_xor = cv2.bitwise_xor(img1, mask)

cv2.imshow('AND', mask_and)
cv2.imshow('OR', mask_or)
cv2.imshow('XOR', mask_xor)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# cv2.adaptiveThreghold(img--원본이미지, vlaue--조건을 만족할 픽셀에 적용할 최대값,
# method--임계치를 적용하는 법, thresholdType, blocksize--블록크기, c--임계값의 보정값)
#                                 ㄴ "ADAPTIVE_THRESH_MEAN_C": 픽셀 중심으로 블럭사이즈x블럭사이즈 구역내의
#                               픽셀 평균값 - C 를 임계치로 함,,,
#                                    "ADPATIVE_THRESH_GAUSSIN_C": 픽셀 중심으로 블럭사이즈x블럭사이즈 구역내의
#                               픽셀에 가우스 가중치를 적용한 평균값에서 C를 뺀 값을 임계치로 함, 현재 검사하는 픽셀에서 가까운 픽셀의 가중치가 크다.
import cv2

img1 = cv2.imread('./data/img1.png', 0) # gray scale
img2 = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize = 9, C = 0)

cv2.imshow('Edge Detection', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
