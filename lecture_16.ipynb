최소 제곱 표현

```python
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
mx = np.mean(x)
my = np.mean(y)
print('mean of x: {}, mean of y: {}'.format(mx, my))

down = sum([(i - mx)**2 for i in x]) #sum([a, b, c])

top = 0
for i in range(len(x)):
    top = top + (x[i] - mx) * (y[i] - my)

a = top / down
b = my - (mx * a)
print('regression line: y = {}x + {}'.format(a, b))
```

MSE 구현

```python
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

a = 3
b = 76

y_pred = []
for i in range(len(x)):
    y_pred.append(a * x[i] + b)

def mse(y, y_pred):
    return (1/len(x)) * sum((y - y_pred)**2)

print('x:', x)
print('y:', y)
print('y_pred:', y_pred)
print('mse:', str(mse(y, y_pred)))
```

```python
# 선형회기 구현
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

a = 0
b = 0
lr = 0.03  # learning rate
epochs = 2001 # epochs
n = len(x) # number of x

# gradient decent
for i in range(epochs):
    y_pred = a * x + b
    error = y - y_pred

    a_diff = (2/n) * sum(-x * error)
    b_diff = (2/n) * sum(-error)
    a = a - lr * a_diff
    b = b - lr * b_diff
    if(i % 100 == 0):
        print('epoch: {}, a: {}, b: {}'.format(i, a, b))

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```

다중 선형 회기

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter3D(x1, x2, y)
plt.title("Scatter plot")
plt.show()

a1 = 0
a2 = 0
b = 0
lr = 0.005 # learning rate
epochs = 10001 # epochs
n = len(x1) # number of x

# gradient decent
for i in range(epochs):
    y_pred = a1 * x1 + a2 * x2 + b
    error = y - y_pred
    a1_diff = (2/n) * sum(-x1 * error)
    a2_diff = (2/n) * sum(-x2 * error)
    b_diff = (2/n) * sum(-error)
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    if(i % 1000 == 0):
      print('epoch: {}, a1: {}, a2: {}, b: {}'.format(i, a1, a2, b))
    
y_pred = a1 * x1 + a2 * x2 + b 
print('y:', y)
print('y_pred:', y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

model = Sequential()
model.add(Dense(1, input_dim = 1, activation = 'linear'))

model.compile(optimizer = 'sgd', loss = 'mse')
model.fit(x, y, epochs = 2000)

y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()

#hours = int(input("Input hours as integer"))
hours = 7
exp_score = model.predict([hours])
print('The expected score is', exp_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])

model = Sequential()
model.add(Dense(1, input_dim = 2, activation = 'linear'))

model.compile(optimizer = 'sgd', loss = 'mse')
model.fit(x, y, epochs = 2000)

y_pred = model.predict(x)

hours = 7
tutor = 4
exp_score = model.predict([[hours, tutor]])
print('The expected score is', exp_score)
```
