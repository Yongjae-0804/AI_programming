### 워드 클라우드
각 단어의 빈도수를 시각화

```python
import wordcloud
import wikipedia
import matplotlib.pyplot as plt

kbo = wikipedia.page('KBO League')
txt_kbo = kbo.content
print(txt_kbo)
image = wordcloud.WordCloud(width = 1000, height = 700).generate(txt_kbo)

plt.figure(figsize = (40,30))
plt.imshow(image)
plt.show()
```

```python
kbo = wikipedia.page("KBO League")
txt_kbo = kbo.content
print(txt_kbo)
s_words = wordcloud.STOPWORDS.union({'one', 'two', 'three', 'year', 'first', 'second'})
image = wordcloud.WordCloud(width = 1000, height = 700, stopwords =
 s_words).generate(txt_kbo)
plt.figure(figsize = (40, 30))
plt.imshow(image)
plt.show()
```
