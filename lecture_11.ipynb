### matplotlib

```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3, 35004]

plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')
plt.title("GDP per capita")
plt.xlabel("years")
plt.ylabel("dollars")
#plt.savefig("gdp_per_capita.png", dpi = 600)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_num = 100
x = [i for i in range(0, x_num)]
y = [np.random.exponential(10) for k in range(0, x_num)]

plt.title("random number")
plt.plot(x,y, color = 'r')
plt.xlabel('x')
plt.ylabel('random number')
plt.show()
```

```python
x = [x for x in range(-20,20)]
y1 = [2*i for i in x]
y2 = [i**2 - 4*i + 3 for i in x]
y3 = [-i**2 -4*i -3 for i in x]

plt.title('three graph')
plt.plot(x,y1,'r-', x,y2,'g--', x,y3,'b*:')
plt.axis([-30,30,-40,40])
plt.show()
```

```python
x = [x for x in range(-20, 20)]
y1 = [2 * i for i in x]
y2 = [i**2 - 4 * i + 3 for i in x]
y3 = [-i**2 - 4 * i - 3 for i in x]
plt.title("Three graphs together")

#plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.plot(x, y1, 'r--', label = 'y=2x')
plt.plot(x, y2, 'g^-', label = 'y=x^2-4x+3')
plt.plot(x, y3, 'b*:', label = 'y=-x^2-4x-3')

plt.axis([-30, 30, -40, 40])
plt.legend()
plt.show()
```

```python
import math

x = []
y = []
z = [] #넘파이 배열도 사용가능

for i in range(360):
    x.append(i)
    y.append(math.sin(math.radians(i))) # 각도를 레디안으로 변경해주는 매서드 사용
    z.append(math.cos(math.radians(i)))

plt.title('sinusoidal')
plt.plot(x, y, color = 'g', label = 'sin')
plt.plot(x,z, color = 'r', label = 'cos')
plt.legend()
plt.xlabel('radian')
plt.show()
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020] #막대가 8개다.
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3, 35004] # 각각의 막대의 값

plt.title("Bar exercise")
plt.bar(range(len(years)), gdp, label = 'gdp per capita') # 막대그래프는 bar를 사용
plt.legend()
plt.xlabel("Year")
plt.ylabel("USD")
plt.xticks(range(len(years)), years) # plt.xticks(개수, 숫자)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_bins = 200
gauss1 = np.random.randn(100000000)
gauss2 = np.random.randn(100000000) + 3

plt.title('Two Normal distribution')
plt.hist(gauss1, n_bins, color = 'blue', alpha = 0.5) #alpha : 투명도, 투명도를 주면 색이 합쳐진다.
plt.hist(gauss2, n_bins, color = 'red', alpha = 0.5) # 없으면 덮어 그리기때문에 안 보이는 부분 발생.
plt.show()
```

```python
# 서브 플롯 (한번에 그래프 여거개 그리기)
import numpy as np
import matplotlib.pyplot as plt

fig, sp = plt.subplots(2, 2, figsize = (10, 10)) #plt.subplots(행 , 열, 크기)
sp[0, 0].plot([1, 2, 3, 4, 5], 'r')
sp[1, 0].plot([2, 3, 2, 3, 1], 'b')
sp[0, 1].plot([5, 5, 5, 2, 1], 'g')
sp[1, 1].plot([5, 4, 3, 2, 1], 'y')

plt.show()
```

```python
# 좀더 좋은 방법
fig = plt.figure()

sp1 = fig.add_subplot(2, 2, 1) # 2 by 2 중 첫번 째 순서는 좌-우, 상-하
sp2 = fig.add_subplot(2, 2, 2)
sp3 = fig.add_subplot(2, 2, 3)
sp4 = fig.add_subplot(2, 2, 4)

sp1.plot([1, 2, 3, 4, 5], 'r')
sp2.plot([2, 3, 2, 3, 1], 'b')
sp3.plot([5, 5, 5, 2, 1], 'g')
sp4.plot([5, 4, 3, 2, 1], 'y')

plt.show()
```
